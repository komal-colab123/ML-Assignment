```python
import matplotlib.pyplot as plt

fuel_types = ['Diesel', 'Petrol', 'CNG', 'LPG', 'Electric']
counts = [3852, 3325, 62, 12, 2]

plt.bar(fuel_types, counts, color='skyblue')

plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Fuel Type Distribution')

plt.show()
```

```python
data['Transmission'].value_counts()
```

```python
import matplotlib.pyplot as plt

transmission_counts = data['Transmission'].value_counts()

transmission_counts.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(6,6))

plt.title('Transmission Type Distribution')

plt.ylabel('')

plt.show()
```

```python
plt.hist(data['Mileage'].dropna(), bins=30, color='skyblue')
plt.title('Mileage Distribution')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()
```
