```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df_sal = pd.read_csv(r'C:\Users\komal\OneDrive\Desktop\Salary_Data.csv')
```

```python
df_sal.head()
```

```python
df_sal.describe()
```

```python
plt.title('Salary Distribution Plot')  
sns.histplot(df_sal['Salary'], kde=True)  
plt.show()
```

```python
plt.scatter(df_sal['YearsExperience'], df_sal['Salary'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = df_sal.iloc[:, :1]  
y = df_sal.iloc[:, 1:]  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     
y_pred_train = regressor.predict(X_train) 
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error (Test Set):", mse)
```

```python
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)

print("Correct RMSE:", rmse)
```

```python
mae = mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error (Test Set):", mae)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred_test)
print("R Score (Test Set):", r2)
```
